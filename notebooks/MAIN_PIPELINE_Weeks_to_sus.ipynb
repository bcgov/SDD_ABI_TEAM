```python
def update_sustainment_prediction_table():   
    #this function works very similarly to the main pipeline, except it gets the 
    #case_numbers from the sustainment_predictions table that are predicted to have 
    #sustainment, then passes into the preprocessing function only those case numbers that don't already exist within the weeks_to_sustainment
    #table already
    import psycopg2
    import os
    import pandas as pd
    import time    
    from psycopg2.errorcodes import UNIQUE_VIOLATION
    from psycopg2 import errors

    import datetime

    import psycopg2
    #connect to openshift database
    conn = psycopg2.connect(database=os.environ['db_name'], user=os.environ['db_username'], password=os.environ['db_password'], host=os.environ['db_host'], port= os.environ['db_port'])
    cursor = conn.cursor()
    
    #get case numbers that are predicted to get sustainment
    query = 'SELECT case_number FROM sustainment_predictions where prediction = 1'
    
    df = pd.read_sql(query, con=conn)
    #put these case numbers into a list
    inpt = df.case_number.tolist()
    
    
    #get case numbers from weeks_to_sustainment_predictions table
    query2 = 'SELECT case_number FROM weeks_to_sustainment_predictions'
    df2 = pd.read_sql(query2, con=conn)
    
    inpt2 = df2.case_number.tolist()
    
    #return list of case numbers that exist in the sustainment predcitions table that 
    #are predicted to get sustainable employment, but aren't already in the 
    #weeks to sustainment table
    inpt3 = list(set(inpt) - set(inpt2))
    
    
    #enter this list as an input
    p = update_weeks_to_sus(inpt3)

    print(p.shape)
    g=p

    if g.empty != True:
        import pickle
        # load the dimension reduction and ML model
        pca = pickle.load(open('/home/jovyan/work/pca_weeks_to_sus.pkl', 'rb'))
        case_nums = list(g.Case_Number)
        #reduce the dimensions
        x_pred  = pca.transform(g.drop(columns = ['Case_Number']))
        model = pickle.load(open('/home/jovyan/work/finalized_weeks_to_sus_model.sav', 'rb'))
        
        #run the prediction
        result = model.predict(x_pred)


        #get current time to write into the 'prediction time' field in the DB
        t = time.strftime('%Y-%m-%d %H:%M:%S')
        #connect to the db
        conn = psycopg2.connect(database=os.environ['db_name'], user=os.environ['db_username'], password=os.environ['db_password'], host=os.environ['db_host'], port= os.environ['db_port'])
        cursor = conn.cursor()
        query = 'SELECT * FROM weeks_to_sustainment_predictions'
        
        df = pd.read_sql(query, con=conn)
        # Fetch a single row using fetchone() method.
        print(df)
        
        #write out how many results are being updated
        po = 'updating' + str(len(result)) + 'results'
        os.write(1, po.encode())

        
        conn = psycopg2.connect(database=os.environ['db_name'], user=os.environ['db_username'], password=os.environ['db_password'], host=os.environ['db_host'], port= os.environ['db_port'])
        
        #for each result update the table
        for n in range(len(result)):

            cursor = conn.cursor()
            #query to update the table
            sql = """INSERT INTO weeks_to_sustainment_predictions (case_number, prediction, prediction_time, prediction_type) 
            VALUES ('"""+str(case_nums[n]) +"""','"""+ str(round(result[n]))+"""', '"""+t+ """', '"""+'normal_prediction' +"""') ON CONFLICT(case_number) DO UPDATE SET prediction = '""" + str(round(result[n])) + """',prediction_time='""" + t + """', prediction_type = 'normal_prediction';"""

            try:
                cursor.execute(sql)
                # commit the changes to the database
                conn.commit()
            #ignore conlifcts of unique violation    
            except errors.lookup(UNIQUE_VIOLATION) as e:
                os.write(1, e)
                continue
        conn.close()
        conn = psycopg2.connect(database=os.environ['db_name'], user=os.environ['db_username'], password=os.environ['db_password'], host=os.environ['db_host'], port= os.environ['db_port'])
        cursor = conn.cursor()
        
        #check that the results have been updated from the print out
        query = 'SELECT * FROM weeks_to_sustainment_predictions'

        df = pd.read_sql(query, con=conn)
        print(df)
    else:
        os.write(1, 'no data to update'.encode())
        print('no data to update')
update_sustainment_prediction_table()
```
