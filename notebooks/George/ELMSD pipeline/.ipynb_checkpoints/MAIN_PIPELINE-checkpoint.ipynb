```python
def update_sustainment_prediction_table():   
    #import this is the function that is the actual pipeline
    #it first reads all the case numbers from the database on openshift, in order
    #to have a list of case numbers that have already been predicted on
    #it then feeds this list into the preprocessing function, to return a dataframe of 
    #case numbers that need to be predicted on
    #the predictions are made, and then they are uploaded back into the database
    import psycopg2
    import os
    import pandas as pd
    import time    
    from psycopg2.errorcodes import UNIQUE_VIOLATION
    from psycopg2 import errors

    import datetime

    import psycopg2
    #connect to the OpenShift database with the predictions (using environmental variables)
    conn = psycopg2.connect(database=os.environ['db_name'], user=os.environ['db_username'], password=os.environ['db_password'], host=os.environ['db_host'], port= os.environ['db_port'])
    cursor = conn.cursor()
    
    #get the case numbers that already exist in the data table
    query = 'SELECT case_number FROM sustainment_predictions'
    
    #turn those case numbers into a list
    df = pd.read_sql(query, con=conn)

    inpt = df.case_number.tolist()
    
    #that list is now the input of the preprocessing function (described above)
    p = preprocess_cna_data_by_case_num(inpt)

    print(p.shape)
    g=p
    #the variable g is the dataframe of case numbers that need to be predicted on
    #if g isn't empty then run the predictions, or else finish the function
    if g.empty != True:
        import pickle
        #read in the pca model
        pca = pickle.load(open('/home/jovyan/work/pca_2.pkl', 'rb'))
        #seperate out the case numbers (as they can't be predicted on)
        case_nums = list(g.Case_Number)
        #run the dataframe through the pca model to reduce the number of dimensions
        x_pred  = pca.transform(g.drop(columns = ['Case_Number']))
        #load the nearest neighbors model
        model = pickle.load(open('/home/jovyan/work/finalized_model_2.sav', 'rb'))
        #run the prediction on the model
        result = model.predict(x_pred)
        #import the other nearest neighbors model to get the closest case number
        nn = pickle.load(open('/home/jovyan/work/nearest_neighbors.pkl', 'rb'))
        cn = pickle.load(open('/home/jovyan/work/nearest_neighbors_cns.pkl', 'rb'))

        #get the current time
        t = time.strftime('%Y-%m-%d %H:%M:%S')
        
        #connect back to the database
        conn = psycopg2.connect(database=os.environ['db_name'], user=os.environ['db_username'], password=os.environ['db_password'], host=os.environ['db_host'], port= os.environ['db_port'])
        cursor = conn.cursor()
        query = 'SELECT * FROM sustainment_predictions'
        df = pd.read_sql(query, con=conn)
        # Fetch a single row using fetchone() method.
        print(df)
        
        #print output
        po = 'updating' + str(len(result)) + 'results'
        os.write(1, po.encode())
        
        conn = psycopg2.connect(database=os.environ['db_name'], user=os.environ['db_username'], password=os.environ['db_password'], host=os.environ['db_host'], port= os.environ['db_port'])
        for n in range(len(result)):
            #get closest case numbers for each case
            f = nn.kneighbors(g.drop(columns = ['Case_Number']).iloc[n].values.reshape(1, -1), return_distance=False)
            l = []
            #make nearest neighbors into a list
            for x in f[0]:    
                l.append(cn[x])
           
            cursor = conn.cursor()
            #update the table
            sql = """INSERT INTO sustainment_predictions (case_number, prediction, prediction_time, prediction_type, closest_case_numbers) 
            VALUES ('"""+case_nums[n] +"""','"""+ result[n]+"""', '"""+t+ """', '"""+'normal_prediction' +"""',"""+str(tuple(l[1:]))+""") ON CONFLICT(case_number) DO UPDATE SET prediction = '""" + result[n] + """',prediction_time='""" + t + """', prediction_type = 'normal_prediction';"""
            #case_number = '""" + case_nums[n] + """',
            #sql = """UPDATE sustainment_predictions SET case_number = """ + case_nums[n] + """, prediction = """ + result[n] + """,prediction_time='""" + t + """', prediction_type = 'normal_prediction' """
            
            os.write(1, sql.encode())
            #do the update
            try:
                cursor.execute(sql)
                conn.commit()
            #if the case number has already been updated, ignore error ad continue
            except errors.lookup(UNIQUE_VIOLATION) as e:
                os.write(1, e)
                continue
        conn.close()
        #check that update is complete
        conn = psycopg2.connect(database=os.environ['db_name'], user=os.environ['db_username'], password=os.environ['db_password'], host=os.environ['db_host'], port= os.environ['db_port'])
        cursor = conn.cursor()
        query = 'SELECT * FROM sustainment_predictions'
        
        df = pd.read_sql(query, con=conn)
        print(df)
    else:
        os.write(1, 'no data to update'.encode())
        print('no data to update')
update_sustainment_prediction_table()
```
